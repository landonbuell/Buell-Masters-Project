```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
# Load in the CIFAR10 dataset
(X_train,y_train),(X_test,y_test) = tf.keras.datasets.cifar10.load_data()

print("X train shape: {0}".format(X_train.shape))
print("y train shape: {0}".format(y_train.shape))
print("X test shape: {0}".format(X_test.shape))
print("y test shape: {0}".format(y_test.shape))
```

```python
# Store som important varibles
trainDataSize = y_train.shape[0]
testDataSize = y_test.shape[0]
batchSize = 64
```

```python
# Define the names of each class
CLASS_NAMES = ["Plane","Car","Bird","Cat","Deer","Dog","Frog","Horse","Ship","Truck"]
```

```python
plt.figure(figsize=(10,10))
for ii in range(25):
    plt.subplot(5,5,ii+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(X_test[ii], cmap=plt.cm.binary)
    plt.xlabel(CLASS_NAMES[y_test[ii,0]])
plt.show()
```

```python
def oneHotEncode(labels,numClasses):
    """ One-Hote encode a vector of samples """
    oneHot = np.zeros(shape=(labels.size,numClasses),dtype=np.uint8)
    for ii,tgt in enumerate(labels):
        oneHot[ii,tgt] = 1
    return oneHot
```

```python
def processImages(images):
    """ Function to preprocess a group of images """
    images = (images / 255.0) - 0.5
    return images
```

```python
# Apply the processing to all images
trainImages = processImages(X_train)
testImages = processImages(X_test)
```

```python
# Bundle up the data into batches
trainDataset = tf.data.Dataset.from_tensor_slices(trainImages).shuffle()
testDataset = tf.data.Dataset.from_tensor_slices(testImages)
```
